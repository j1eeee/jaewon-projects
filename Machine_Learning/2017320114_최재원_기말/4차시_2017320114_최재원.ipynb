## **퍼셉트론이란?**

퍼셉트론은 신경망의 가장 기초가 되는 기본 단위입니다.

자세히 보면 "퍼셉트론(Perceptron)"이라는 단어는 두 단어의 조합임을 알 수 있습니다.

* Perception 무엇인가를 감지할 수 있는 능력

따라서 퍼셉트론은 생물학적 뉴런의 작업을 시뮬레이션하는 인공 뉴런이라고 할 수 있습니다.


## **퍼셉트론(Perceptron)의 구성 요소**

퍼셉트론은 간단한 결정을 내릴 수 있는 인공 뉴런이라고 할 수 있습니다.

퍼셉트론은 3가지 구성요소로 이루어져 있습니다.

* 입력(Inputs): 각각의 입력 값은 feature에 대응합니다. 

* 가중치(Weights): 각 입력은, 중요도를 나타내는 가중치를 가지고 있습니다. 입력의 가중치가 크다는 것은, 결정을 내림에 있어서 해당 입력값이 중요한 역할을 하고 있다는 것을 의미합니다.

* 출력(Output): 마지막으로, 퍼셉트론은 입력과 가중치를 사용하여 출력을 생성합니다. 출력 유형은 문제의 특성 및 데이터 도메인에 따라 다릅니다. 참과 거짓을 해결해야하는 문제(Classification)라면 0 또는 1로 출력값을 나타내야할 것이고, 만약 결과값이 연속된 수치(Regression)라면 정수로 결과값을 표현할 수 있어야할 것입니다.

## **가중합(Weighted Sum) 구하기**

입력과 가중치를 출력으로 변환하기 위해서는, 우선 각 입력에 대한 가중치를 더해서 가중합을 계산해야합니다.

가중합은 다음과 같은 수식으로 간단히 표현할 수 있습니다.

```
weighted sum = x_1 * w_1 + x_2 * w_2 + ... + x_n * w_n
```

위 수식을 구현한다고 하면 다음의 방식을 통해 계산할 수 있을 것입니다.

1. 가중합을 0으로 초기화합니다.
2. 첫 번째 입력에 해당 가중치를 곱하십시오. 이 결과를 가중합에 더해줍니다.
3. 다음 입력에 해당 가중치를 곱하십시오. 이 결과를 가중합에 더해줍니다.
4. 모든 입력에 대하여 위의 과정을 반복합니다.


## **Practice 2**

```weighted_sum``` 함수를 구현해봅시다.

1. 가중치 합의 값을 저장해 줄 ```weighted_sum```이라는 변수를 생성하고 0으로 초기화해줍시다.

    최종적으로 ```weighted_sum```은 해당 함수에서 반환해주어야 합니다.

2. 입력과 가중치의 쌍을 탐색한 후, 입력과 가중치를 곱해서 가중합에 더해줍시다.

````
class Perceptron:
  def __init__(self, num_inputs=2, weights=[2,1]):
    self.num_inputs = num_inputs
    self.weights = weights
  
  # weighted_sum 메소드 구현하기
  def weighted_sum(self, inputs):
    # 가중합을 저장하는 변수를 생성해주고 초기화해줍니다.
    weighted_sum = 0
    for i in range(self.num_inputs):
      weighted_sum += inputs[i] * self.weights[i]
    # 가중합의 값을 함수의 결과로 반환해줍시다.
    return weighted_sum
````

````
# Perceptron 클래스 외부에서 Perceptron 객체를 생성한 후 주어진 입력([24, 55])에 대한 가중치 합을 출력해봅시다.
cool_perceptron = Perceptron()
print(cool_perceptron.weighted_sum([24, 55]))
````

## **활성화 함수(Activation Function)**

퍼셉트론이 출력을 생성하기 위해서, 앞에서 계산한 가중합으로부터 결과값을 생성해줘야합니다.

결과값을 생성해주는 함수를 활성화 함수라고 합니다.

다양한 활성화 함수들이 존재하는데, 이번 실습에서는 Step Activation Function을 구현해보도록 하겠습니다. 구현할 Step Activation Function은 다음과 같이 작동합니다.

* 입력에 대한 퍼셉트론의 가중합이 양수라면 1을 반환합니다.
* 입력에 대한 퍼셉트론의 가중합이 음수라면 -1을 반환합니다.

## **Practice 3**

앞서 언급한 명세대로 활성화 함수를 구현해봅시다.

1. ```activation()``` 메서드 내에서 ```weighted_sum```이 0보다 크거나 같으면 1을 반환합니다.

2. ```activation()``` 메서드 내에서 ```weighted_sum```이 0보다 작으면 -1을 반환합니다.

````
class Perceptron:
  def __init__(self, num_inputs=2, weights=[1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
    
  def weighted_sum(self, inputs):
    weighted_sum = 0
    for i in range(self.num_inputs):
      weighted_sum += self.weights[i]*inputs[i]
    return weighted_sum
  
  def activation(self, weighted_sum):
    # activation 메소드 구현하기
    if weighted_sum >=0: 
      return 1
    else: 
      return -1
````

## **퍼셉트론 학습시키기**

이제 퍼셉트론은 주어진 입력 값을 예측할 수 있게 설계되었습니다. 퍼셉트론의 또 다른 강점은, 스스로 정답을 향해 학습을 수행한다는 것입니다.

지금 구현한 퍼셉트론은 아무것도 가르쳐주지 않았기 때문에 정확한 분류를 기대할 수 없습니다. 그렇기 위해서 우리는 퍼셉트론에 데이터를 제공하여 정답으로 나아갈 수 있도록 훈련시켜야 합니다.

실제 대부분의 기계학습에서 사용하는 데이터는 tabular 형태를 띄고 있지만, 이번 실습에서는 입력값과 출력값을 dictionary 형태로 표현한 데이터셋을 사용하도록 하겠습니다. 우리가 사용할 데이터셋은 다음과 같은 형태를 지닙니다.

```
training_set = {(18, 49): -1, (2, 17): 1, (24, 35): -1, (14, 26): 1, (17, 34): -1}
```

다음 코드는 우리가 퍼셉트론에 학습시킬 데이터를 생성시켜주는 코드입니다. 다음 코드를 실행시키고 해당 데이터의 특징을 살펴보도록 합시다.

````
import random

def generate_training_set(num_points):
	x_coordinates = [random.randint(0, 50) for i in range(num_points)]
	y_coordinates = [random.randint(0, 50) for i in range(num_points)]
	training_set = dict()
	for x, y in zip(x_coordinates, y_coordinates):
		if x <= 45-y:
			training_set[(x,y)] = 1
		elif x > 45-y:
			training_set[(x,y)] = -1
	return training_set
````

````
#training 데이터셋을 generate_training_set 함수를 사용하여 30개의 포인트를 생성합니다
training_set = generate_training_set(30)
````

````
import seaborn
import matplotlib.pyplot as plt

x_plus = []
y_plus = []
x_minus = []
y_minus = []

for data in training_set:
	if training_set[data] == 1:
		x_plus.append(data[0])
		y_plus.append(data[1])
	elif training_set[data] == -1:
		x_minus.append(data[0])
		y_minus.append(data[1])
    
fig = plt.figure()
ax = plt.axes(xlim=(-25, 75), ylim=(-25, 75))

plt.scatter(x_plus, y_plus, marker = '+', c = 'green', s = 128, linewidth = 2)
plt.scatter(x_minus, y_minus, marker = '_', c = 'red', s = 128, linewidth = 2)

plt.title("Training Set")

plt.show()
````

## **Practice 4**

각 instance에 대하여 오차를 계산하는 ```training()```메소드를 구현해봅시다.

1. ```training()``` 메서드에서 ```training_set```의 각 입력(single instance)에 대한 퍼셉트론의 오류를 계산해봅니다.

    우선, 각 instance에 대한 퍼셉트론의 출력값을 예측해야 합니다. ```for``` 루프 내에서 ```predict```라는 변수를 만들고 ```.activation()```, ```.weighted_sum()``` 메서드를 활용해서 perceptron의 예측값을 해당 변수에 저장하도록 합시다.

2. 학습 데이터의 타겟값을 저장하는 변수 ```actual```을 생성한 후 값을 저장해주도록 합시다.

3. 변수 ```error```를 생성한 후에 실제 값과 예측 값의 차이를 계산해서 저장해주도록 합시다.

````
class Perceptron:
  def __init__(self, num_inputs=2, weights=[1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
    
  def weighted_sum(self, inputs):
    weighted_sum = 0
    for i in range(self.num_inputs):
      weighted_sum += self.weights[i]*inputs[i]
    return weighted_sum
  
  def activation(self, weighted_sum):
    if weighted_sum >= 0:
      return 1
    if weighted_sum < 0:
      return -1
    
  def training(self, training_set):
    # training 메소드 구현해보기
    for inputs in training_set:
      prediction = self.activation(self.weighted_sum(inputs))
      actual = training_set[inputs]
      error = actual - prediction
````

## **가중치 수정하기**

앞서 우리는 데이터와 퍼셉트론의 예측값 사이의 오차를 구할 수 있었습니다.

이 오차를 지표로 삼아 우리는 퍼셉트론이 올바른 방향으로 나아갈 수 있도록 학습시킬 수 있습니다.

퍼셉트론의 예측, 즉 출력값은 입력과 가중치 곱의 총합인 가중합을 통해 결정되기 때문에, 퍼셉트론의 예측값을 변경하기 위해서는 출력값을 변경해줘야합니다. 이 중에서 우리가 임의로 변경할 수 있는 변수는 가중치입니다. 

퍼셉트론 학습의 목표는 가중치를 올바른 방향으로 수정해서, 정답을 출력하는 최적의 가중치를 찾는 것입니다.

그렇다면 우리는 어떻게 최적의 가중치를 찾을 수 있을까요? 

최적의 가중치를 찾는 방법 중, 가장 널리 사용되고 있는 것이 바로 경사하강법(Gradient Descent)입니다. 이번에 우리는 Step Activation Function을 사용하기 때문에, 경사하강법 중 Delta Rule을 통해 가중치를 학습시킬 것입니다.

tweaking_the_weights.png

Delta rule은 다음의 간단한 규칙을 통해 가중치를 수정하게 됩니다.

```
weight = weight + learning rate * (error * input)
```

퍼셉트론은 모든 결과값을 정확하게 예측할 때까지 가중치를 계속 조정합니다. 그렇기 때문에 학습이 완료되기 전에 데이터셋을 여러 번 통과해야 할 수도 있습니다.

## **Practice 5**

Delta Rule로 퍼셉트론을 학습시키기 위해 앞서 구현했던 training 메소드를 수정해봅시다.

1.우선 정답을 찾을 때까지 학습을 반복하기 위해 다음의 요소들을 추가해줍시다.

* ```foundLine = False``` (퍼셉트론이 양수 레이블과 음수 레이블을 구분하는 line을 찾았는지 여부를 나타내는 bool 변수)
* ```while not foundLine:``` (라인, 즉 정답을 찾을 때까지 퍼셉트론을 계속 훈련시키는 while 루프)
* ```total_error = 0``` (퍼셉트론이 각 라운드에서 만드는 총 오류를 계산)
* ```total_error += abs(error)``` (각 라운드에서 퍼셉트론이 만드는 총 오류를 업데이트하기 위해)

2. 알고리즘이 오류를 찾지 못하면 퍼셉트론이 모든 데이터 instance에 대해 정답을 올바르게 예측했다는 의미입니다.

      ```for``` 루프의 밖에서, ```total_error```가 ```0```이라면 ```foundLine``` 값을 ```True```로 바꿔줍시다.

3. 각각의 가중치에 대해 Delta Rule을 적용하여 각 가중치 ```self.weights[i]```를 업데이트합니다

    ```weight = weight + (error * inputs)```

주어진 ```small_training_set```을 이용해서 퍼셉트론을 학습시켜봅시다.

```.weights``` 내부 변수를 출력해서 학습된 가중치 수치 또한 직접 확인해보도록 합니다.

````
# training()메소드에 delta rule 추가

class Perceptron:
  def __init__(self, num_inputs=2, weights=[1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
    
  def weighted_sum(self, inputs):
    weighted_sum = 0
    for i in range(self.num_inputs):
      weighted_sum += self.weights[i]*inputs[i]
    return weighted_sum
  
  def activation(self, weighted_sum):
    if weighted_sum >= 0:
      return 1
    if weighted_sum < 0:
      return -1
    
  def training(self, training_set):
    foundLine = False
    while not foundLine:
      total_error = 0
      for inputs in training_set:
        prediction = self.activation(self.weighted_sum(inputs))
        actual = training_set[inputs]
        error = actual - prediction
        total_error += abs(error)

        for i in range(self.num_inputs):
          self.weights[i] += error * inputs[i]

      if total_error == 0:
        foundLine = True
````

````
# 구현된 결과값을 확인해봅니다.
cool_perceptron = Perceptron()
small_training_set = {(0,3):1, (3,0):-1, (0,-3):-1, (-3,0):1}

cool_perceptron.training(small_training_set)
print(cool_perceptron.weights)
````

## **편향치(Bias)**

앞서 우리는 퍼셉트론이 가중치를 조정하여 올바른 출력을 생성하도록 학습 될 수 있음을 알 수 있었습니다.

그러나 퍼셉트론의 정확도를 높이기 위해 약간의 조정이 필요한 경우가 있습니다. 이를 위해 우리는 퍼셉트론에 편향치를 추가합니다.

편향이 추가된 가중합의 식은 다음과 같습니다.

```
weighted sum = x_1 * w_1 + x_2 * w_2 + ... + x_n * w_n​ + w_b
```

다음 두 가지 변경사항만 고려하면 앞서 구현한 퍼셉트론에 편향을 적용할 수 있을 것입니다.

* 입력 데이터의 마지막에 1을 추가합니다 (이제 2 대신 3 개의 입력이 있음).
* 가중치 목록에 편향 가중치를 추가합니다 (이제 2 개 대신 3 개의 가중치가 있음).

````
class Perceptron:
  #  __init__ 생성자를, 편향치가 적용될 수 있도록 적합하게 수정해봅시다.
  def __init__(self, num_inputs=3, weights=[1,1,1]):
    self.num_inputs = num_inputs
    self.weights = weights
    
  def weighted_sum(self, inputs):
    weighted_sum = 0
    for i in range(self.num_inputs):
      weighted_sum += self.weights[i]*inputs[i]
    return weighted_sum
  
  def activation(self, weighted_sum):
    if weighted_sum >= 0:
      return 1
    if weighted_sum < 0:
      return -1
    
  def training(self, training_set):
    foundLine = False
    while not foundLine:
      total_error = 0
      for inputs in training_set:
        prediction = self.activation(self.weighted_sum(inputs))
        actual = training_set[inputs]
        error = actual - prediction
        total_error += abs(error)
        for i in range(self.num_inputs):
          self.weights[i] += error*inputs[i]
      if total_error == 0:
        foundLine = True
````

## **퍼셉트론을 통한 선형 분류**

퍼셉트론은 예상대로 작동하지만 지금은 훈련 과정을 확인할 수 없습니다. 무슨 일이 일어나고 있는지 더 잘 이해하기 위해 퍼셉트론의 훈련 과정을 시각화 해봅니다.

가중치는 훈련 과정 전반에 걸쳐 변경되므로 해당 가중치를 의미있게 시각화 할 수만 있다면 기울기-절편 형식을 사용하여 선을 나타낼 수 있다는 것을 알고있을 것입니다. 퍼셉트론의 가중치는 퍼셉트론이 나타내는 선의 기울기와 절편을 찾는 데 사용할 수 있습니다.

* 기울기 = ```-self.weights[0] /self.weights[1]```
* 절편 = ```-self.weights[2] /self.weights[1]```

다음 코드를 실행시켜 앞서 생성한 퍼셉트론 코드가 제대로 된 것이 맞는지 직접 확인해보도록 합시다.

````
import seaborn
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rc
from IPython.display import HTML
import random
````

````
lines = []

class Perceptron:
  def __init__(self, num_inputs=3, weights=[1,1,1]):
    self.num_inputs = num_inputs
    self.weights = weights

  def weighted_sum(self, inputs):
    weighted_sum = 0
    for i in range(self.num_inputs):
      weighted_sum += self.weights[i]*inputs[i]
    return weighted_sum

  def activation(self, weighted_sum):
    if weighted_sum >= 0:
      return 1
    if weighted_sum < 0:
      return -1

  def training(self, training_set):
    foundLine = False
    while not foundLine:
      total_error = 0
      for inputs in training_set:
        prediction = self.activation(self.weighted_sum(inputs))
        actual = training_set[inputs]
        error = actual - prediction
        total_error += abs(error)
        for i in range(self.num_inputs):
          self.weights[i] += error*inputs[i]

      slope = -self.weights[0]/self.weights[1]
      intercept = -self.weights[2]/self.weights[1]
      y1 = (slope * 0) + intercept
      y2 = (slope * 50) + intercept
      lines.append([[0,50], [y1, y2]])

      if total_error == 0:
        foundLine = True
````

````
from matplotlib.colors import ListedColormap

# Perceptron의 decision boundary를 그려주는 함수입니다.
# 함수 출처: http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html
def plot_decision_boundary(classifier, X, y, title):
  xmin, xmax = np.min(X[:, 0]) - 0.05, np.max(X[:, 0]) + 0.05
  ymin, ymax = np.min(X[:, 1]) - 0.05, np.max(X[:, 1]) + 0.05
  step = 0.01
  cm = plt.cm.coolwarm_r
  thr = 0.0
  xx, yy = np.meshgrid(np.arange(xmin - thr, xmax + thr, step), np.arange(ymin - thr, ymax + thr, step))
  if hasattr(classifier, 'decision_function'):
    Z = classifier.decision_function(np.hstack((xx.ravel()[:, np.newaxis], yy.ravel()[:, np.newaxis])))
  else:
    Z = classifier.predict_proba(np.hstack((xx.ravel()[:, np.newaxis], yy.ravel()[:, np.newaxis])))[:, 1]
  Z = Z.reshape(xx.shape)
  plt.contourf(xx, yy, Z, cmap=cm, alpha=0.8)
  plt.colorbar()
  plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(['#FF0000', '#0000FF']), alpha=0.6)
  plt.xlim(xmin, xmax)
  plt.ylim(ymin, ymax)
  plt.xticks((0.0, 1.0))
  plt.yticks((0.0, 1.0))
  plt.title(title)
````

````
df = pd.read_csv(data_url)
````

````
#1. 데이터 전처리

#feature column 고르기
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Survived']]

#NA가 있는 열 drop하기
df = df.dropna()

#'Sex'feature 정수형으로 인코딩
le = LabelEncoder()
df.Sex = le.fit_transform(df.Sex)

# Embarked feature one-hot encoding
def onehot(data, feature):
  '''
  data의 feature column을 one hot으로 변환해줍니다.
  data: pandas DataFrame
  feature: string, 데이터 프레임의 column 이름
  '''
  return pd.concat([data, pd.get_dummies(data[feature], prefix=feature)], axis=1).drop([feature], axis=1)

df = onehot(df, 'Embarked')

# x와 y를 분리 (lable은 'Survived Feature)
y = df[['Survived']].to_numpy().ravel()
x = df.drop(columns=['Survived'])

#train data, test data 나누기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .3)
````

- 다음은 4-1 과정을 통해서 각각의 파라미터들에 대한 모델을 생성 후 각각의 파라미터들에 대한 최적 값을 그 파라미터의 특징에 맞게 조절해 나갔다. 예를 들면, hidden layer size 같은 경우에는 default 값인 100으로부터 시작하여 100씩 높여가며 결과를 확인했으며, max_iter 같은 경우에는 1000씩, learning rate 은 0.001씩 그리고 activation은 sklearn에서 제공된 activation 종류를 하나씩 해보았다. 이렇게 fitting 된 모델을 4-2 과정을 통해서 비교를 해가며 더 높은 정확도를 찾아갔다.

- 이 과정에서 hidden_layer나 max_iter 같은 경우에는 파라미터 값을 너무 높이면 overfitting 되어서 오히려 정확도가 떨어지는 것을 확인하였고, 어떠한 경우에는 pruning parameter를 조정하지 않은 baseline 모델보다도 낮게 나왔다.

- 결과적으로 hidden_layer_sizes = 200, max_iter = 2000, activation = 'logistic', learning_rate_init = 0.003 로 final model의 파라미터를 조정해 대략 정확도 0.82로 최고의 성능을 찾을 수 있었다.
